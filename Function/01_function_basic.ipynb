```python
def marksCalculator(marks):
    total = 0
    avg_marks =0
    for i in range(len(marks)):
        total = total + marks[i]
    avg_marks = total/len(marks)

    return total,avg_marks
```

```python
marksCalculator([12,12,13,14])[0]=12
```

```python
def sum_all_numbers(*numbers):
    total =0
    for num in numbers:
        total +=num
    return total
```
