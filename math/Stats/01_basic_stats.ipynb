```python
import numpy as np

population = np.random.rand(10,1000,5000) # here rand  parameters are rows, columns, depth
len(population)

mean_population = np.mean(population)
print(mean_population)
```

```python
sample_1 = np.random.choice(population[0][0], size=100)
sample_2 = np.random.choice(population[0][1], size=100)
sample_3 = np.random.choice(population[0][2], size=100)
sample_4 = np.random.choice(population[0][3], size=100)

print(np.mean(sample_1), np.mean(sample_2), np.mean(sample_3), np.mean(sample_4))

# avg of samples
print((np.mean(sample_1) + np.mean(sample_2) + np.mean(sample_3) + np.mean(sample_4)) / 4)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

salary = np.random.exponential(scale=50000, size=1000)

plt.figure(figsize=(10, 6))
plt.hist(salary, bins=30, density=True, alpha=0.6, color='g')
sns.kdeplot(salary, color='k', bw_adjust=0.5)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Density')
plt.show()
```

```python
print('Salary Mean : ', np.mean(salary))
s1 = np.random.choice(salary, size=100)
s2 = np.random.choice(salary, size=100)
s3 = np.random.choice(salary, size=100)
s4 = np.random.choice(salary, size=100)

# avg of samples
print((np.mean(s1) + np.mean(s2) + np.mean(s3) + np.mean(s4)) / 4)
```
