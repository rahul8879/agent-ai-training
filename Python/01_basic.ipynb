```python
# Bitwise operations
a = 10  # 1010 in binary
b = 4   # 0100 in binary
print(a & b)  # Bitwise AND
print(a | b)  # Bitwise OR


# how this works
# Bitwise AND
# 1010
# 0100
# ----
# 0000

# Bitwise OR
# 1010
# 0100
# ----
# 1110

# Bitwise XOR
# 1010
# 0100
# ----
# 1110

# Bitwise NOT
# 1010
# ----
# 0101

# Bitwise left shift
# 1010
# ----
# 10100

# Bitwise right shift
# 1010
# ----
# 0101
```
