```python
# Q) Find the index of a specific element in a list
marks =[24,25,27,28]
key = 25

for i in range(len(marks)):
    if marks[i] == key:
        print("Key found at index", i)
        break
    print(marks[i])


# q) find the occurance of some element in a list
marks =[24,25,27,28,25,25]
key = 25
count = 0
for i in range(len(marks)):
    if marks[i] == key:
        count = count + 1

print("Key found", count, "times")
```

```python
sales = [[29,25,26],[27,28,29],[30,31,32]]
avg_sales = [] # output 

for i in sales:
    sum = 0
    for j in i:
        sum = sum + j
    avg_sales.append(round(sum / len(i),2)) # why round here ? to format the output to 2 decimal places

print(avg_sales)
```
