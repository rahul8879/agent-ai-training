## Task 1: Type Casting and Operations
Implement a function that safely adds a numeric string to an integer.

**Function:**
- `add_numeric_string(num_str: str, value: int) -> float`

**Behavior:**
- Convert `num_str` to a float and add `value`.
- If `num_str` cannot be converted, return `None`.

**Examples:**
- `add_numeric_string("5", 5) -> 10.0`
- `add_numeric_string("5.5", 2) -> 7.5`
- `add_numeric_string("abc", 3) -> None`


```python
def add_numeric_string(num_str: str, value: int) -> float:
    result = round(float(num_str) + value, 2)
    return result

add_numeric_string("3", 2)
```

## Task 3: Keywords and Identifiers
Implement a function that checks whether a given string is a valid Python identifier and not a keyword.

**Function:**
- `is_valid_identifier(name: str) -> bool`

**Behavior:**
- Return `True` only if `name` is a valid identifier and is not a Python keyword.

**Examples:**
- `is_valid_identifier("age1") -> True`
- `is_valid_identifier("1age") -> False`
- `is_valid_identifier("and") -> False`


```python
def is_valid_identifier(name):
    # Check if the name is not empty
    if not name:
        return False
    # Check if the first character is a letter or underscore
    if not (name[0].isalpha() or name[0] == '_'):
        return False
    # Check if all characters are alphanumeric or underscores
    for char in name:
        if not (char.isalnum() or char == '_'):
            return False
    return True

is_valid_identifier('valid_name')  # True
```
