## Task 1: Type Casting and Operations
Implement a function that safely adds a numeric string to an integer.

**Function:**
- `add_numeric_string(num_str: str, value: int) -> float`

**Behavior:**
- Convert `num_str` to a float and add `value`.
- If `num_str` cannot be converted, return `None`.

**Examples:**
- `add_numeric_string("5", 5) -> 10.0`
- `add_numeric_string("5.5", 2) -> 7.5`
- `add_numeric_string("abc", 3) -> None`


```python
def add_numeric_string(num_str: str, value: int) -> float:
     try:
       convert_to_float = float(num_str)   
       return   value+convert_to_float
     except ValueError:
        print ('ValueError {} ', ValueError)
        return None
        
str_val_valid = "5"
int_val = 5

result_valid = add_numeric_string(str_val_valid, int_val)
if result_valid is not None:
     print(f"Result (valid): {result_valid}")

str_val_valid = "5.5"
int_val = 2
result = add_numeric_string(str_val_valid, int_val)
if result is not None:
    print(f"Result (valid): {result}")
else:
    print(f"Result (invalid): {result}")    

int_val = 5
str_val_invalid = "abc"
result_invalid = add_numeric_string(str_val_invalid, int_val)
print(f"Result (invalid): {result_invalid}")    
```

## Task 3: Keywords and Identifiers
Implement a function that checks whether a given string is a valid Python identifier and not a keyword.

**Function:**
- `is_valid_identifier(name: str) -> bool`

**Behavior:**
- Return `True` only if `name` is a valid identifier and is not a Python keyword.

**Examples:**
- `is_valid_identifier("age1") -> True`
- `is_valid_identifier("1age") -> False`
- `is_valid_identifier("and") -> False`


```python
def is_valid_identifier(name):
    # Check if the name is not empty
    if not name:
        return False
    # Check if the first character is a letter or underscore
    if not (name[0].isalpha() or name[0] == '_'):
        return False
    # Check if all characters are alphanumeric or underscores
    for char in name:
        if not (char.isalnum() or char == '_'):
            return False
    return True

print ( ' test 1 {} ',is_valid_identifier('valid_name') ) # True
print ( ' is age1 valid identifier {} ',is_valid_identifier('age1') ) # True
print ( ' is 1age valid identifier {} ',is_valid_identifier('1age') ) # False
print ( ' is and valid identifier {} ',is_valid_identifier('and') ) # True


```

## Task 4: Types and Identity
Implement a function that compares both value and identity of two objects.

**Function:**
- `compare_value_and_id(a, b) -> tuple`

**Behavior:**
- Return a tuple `(same_value, same_identity)` where:
  - `same_value` is `True` if `a == b`.
  - `same_identity` is `True` if `a is b`.

**Example:**
- `compare_value_and_id("10", "10") -> (True, True)` (may vary for strings)
- `compare_value_and_id([1,2], [1,2]) -> (True, False)`


```python
def compare_value_and_id(a, b) -> tuple:
    return (a==b) and ( a is b)
    
print ('is 10 == 10 {} ', compare_value_and_id(10, 10))
print ('is [1,2] = [1,2]{} ', compare_value_and_id([1,2], [1,2]))
```
