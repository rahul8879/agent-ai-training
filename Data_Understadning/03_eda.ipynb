```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['class'] = iris.target
data.head()
```

```python
def label_mapping(x):
    if x==0:
        return 'setosa'
    elif x==1:
        return 'versicolor'
    elif x==2:
        return 'virginica'
    

data['class']= data['class'].apply(label_mapping)
data.head()
```

```python
data.rename({
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width',
    'class': 'species'
}, axis=1, inplace=True)

data.head()
```

```python
data.shape
```

```python
data.plot(kind = 'scatter', x='sepal_length', y='sepal_width')
```

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species')
```

## Conclusion
- 1. The dataset contains 150 samples of iris flowers, each with 4 features (sepal length, sepal width, petal length, petal width) and a target class (species).
- 2. The exploratory data analysis (EDA) revealed distinct separations between the species based on the features, particularly in the petal dimensions.
- 3. Visualizations such as scatter plots and pair plots helped to identify these patterns and relationships in the data.

```python
sns.scatterplot(data=data, x='petal_length', y='petal_width', hue='species')
```

```python
sns.pairplot(data, hue='species')
```

```python
# 1 D analysis : univariate analysis
sns.histplot(data=data[data.species == 'setosa'], x='sepal_length', hue='species',kde=True)
```
