```python
data = {
    'age': np.random.randint(0, 100, 500),
    'income': np.random.randint(20000, 100000, 500),
    'price': np.random.randint(5, 500, 500),
    'city': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], 500),
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Books', 'Furniture'], 500)
}


sales_df = pd.DataFrame(data)

sales_df.head(2)
```

```python
sales_df.groupby('city').agg({
    'income': ['max', 'min', 'mean'],
    'price': ['max', 'min', 'mean']
})
```

```python
sales_df.groupby(['city', 'product_category']).agg({
    'price': ['max', 'min', 'mean','std']
})
```

```python
# in groupby : agg function : you can also pass 
def range_func(x):
    return x.max() - x.min()

result_df = sales_df.groupby(['city', 'product_category']).agg({
    'price': ['max', 'min', 'mean', range_func]
})

# order by
result_df = result_df.sort_values(by=('price', 'mean'), ascending=False)
result_df.head(10)
```

```python
result_df.to_csv('sales_analysis.csv')
```

```python
# sales table vs product table
sales_df = pd.DataFrame({
    'order_id': [1, 2, 3],
    'product_id': [101, 102, 103],
    'quantity': [2, 1, 5]
})

product_df = pd.DataFrame({
    'product_id': [101, 102, 103],
    'product_name': ['Product A', 'Product B', 'Product C'],
    'price': [10.0, 20.0, 30.0]
} )

```

```python
sales_df.head()
```

```python
product_df.head()
```

```python
final_table = pd.merge(sales_df, product_df[['product_id', 'price']], on='product_id', how='left') # left join
```

```python
final_table.head(2)
final_table['total_price'] = final_table['quantity'] * final_table['price']
```

```python
final_table.total_price.describe()
```

```python
# sometime col name may differ in both dataframe
# ex
sales_df = pd.DataFrame({
    'order_id': [1, 2, 3],
    'product_id': [101, 102, 103],
    'quantity': [2, 1, 5]
})

product_df = pd.DataFrame({
    'id': [101, 102, 103],
    'product_name': ['Product A', 'Product B', 'Product C'],
    'price': [10.0, 20.0, 30.0]
} )


pd.merge(sales_df, product_df[['id', 'price']], left_on='product_id', right_on='id', how='left')
```

```python
sales_df = pd.DataFrame({
    'order_id': [1, 2, 3],
    'product_id': [101, 102, 103],
    'quantity': [2, 1, 5]
})

product_df = pd.DataFrame({
    'product_id': [101, 102, 103],
    'product_name': ['Product A', 'Product B', 'Product C'],
    'price': [10.0, 20.0, 30.0]
} )

sales_df.head()
```

```python
sales_df.join(product_df.set_index('product_id'), on='product_id', how='left')
```

```python
# when to use merge and join ??
# Use merge when you need to combine DataFrames based on a key column(s) and you want to specify
#  how to handle overlapping columns.
# Use join when you want to combine DataFrames based on their index or a key column(s) 
# and you want a simpler syntax.
```
