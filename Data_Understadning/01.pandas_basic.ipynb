```python
data = {
    'age': np.random.randint(0, 100, 500),
    'income': np.random.randint(20000, 100000, 500),
    'price': np.random.randint(5, 500, 500),
    'city': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], 500),
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Books', 'Furniture'], 500)
}


sales_df = pd.DataFrame(data)

sales_df.head(2)
```

```python
sales_df.groupby('city').agg({
    'income': ['max', 'min', 'mean'],
    'price': ['max', 'min', 'mean']
})
```

```python
sales_df.groupby(['city', 'product_category']).agg({
    'price': ['max', 'min', 'mean','std']
})
```

```python
# in groupby : agg function : you can also pass 
def range_func(x):
    return x.max() - x.min()

result_df = sales_df.groupby(['city', 'product_category']).agg({
    'price': ['max', 'min', 'mean', range_func]
})

# order by
result_df = result_df.sort_values(by=('price', 'mean'), ascending=False)
result_df.head(10)
```

```python
result_df.to_csv('sales_analysis.csv')
```

```python
result_df.to_html('sales_analysis.html')
```
