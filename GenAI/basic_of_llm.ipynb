```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(
    model="gpt-4o-mini"
)
```

```python
prompt = """You are an email classification system. Classify the following email into one of these categories: Spam, Important, Newsletter, Personal.

Email: {email} and Provide:
1. Overall Match Score (0-100)
2. Strengths (2-3 points)
3. Gaps/Concerns (2-3 points)
4. Final Recommendation: Strong Yes / Yes / Maybe / No
5. Brief reasoning (2 sentences)"""

model.invoke(prompt.format(email=emails[0])).content

```

```python
from pydantic import BaseModel

class EmailClassification(BaseModel):
    email: str
    category: str
    overall_match_score: int
    strengths: list[str]
    gaps_concerns: list[str]
    final_recommendation: str
    brief_reasoning: str
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
parser = JsonOutputParser(pydantic_object=EmailClassification)


prompt = PromptTemplate(
     template= """You are an email classification system. 
    Classify the following email into one of these categories: Spam, Important, Newsletter, Personal.

    Email: {email}
    {format_instructions}

    """,
    input_variables=["email"],
    partial_variables={"format_instructions": parser.get_format_instructions()}
)

chain = prompt | model | parser
```

```python
chain.invoke({"email": emails[0]})
```

```python
for i in range(len(email_df)):
    raw_output = chain.invoke({"email": emails[i]})
    email_df.loc[i, "category"] = raw_output['category']
    email_df.loc[i, "overall_match_score"] = raw_output['overall_match_score']
    # email_df.loc[i, "strengths"] = raw_output['strengths']
    # email_df.loc[i, "gaps_concerns"] = raw_output['gaps_concerns']
    # email_df.loc[i, "final_recommendation"] = raw_output['final_recommendation']
    # email_df.loc[i, "brief_reasoning"] = raw_output['brief_reasoning']
```
