```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
load_dotenv()
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
call_1 = llm.invoke("I received a billing complaint from John about invoice #1042. ")
call_1.content
```

```python
call_2 = llm.invoke(
    "What category does his complaint fall under?")

call_2
```

```python
# manualy context -- history -- LLM
user_turn_1 = "I received a billing complaint from John about invoice #1042. He was charged $500 instead of $250."
ai_turn_1   = llm.invoke(user_turn_1).content

print("TURN 1")
print(f"User: {user_turn_1}")
print(f"AI:   {ai_turn_1}")
print()
```

```python
user_turn_2 = "What category does his complaint fall under?"

stitched_prompt_turn_2 = f"""\
Previous conversation:
User: {user_turn_1}
AI: {ai_turn_1}

Now answer:
User: {user_turn_2}"""

ai_turn_2 = llm.invoke(stitched_prompt_turn_2).content

print(f"User: {user_turn_2}")
print(f"AI:   {ai_turn_2}")
```

```python
# chain == ? AIMessage , Ask ==> HumanMessage
email_questions = [
    "I received a billing complaint from John about invoice #1042. He was charged $500 instead of $250.",
    "What category does his complaint fall under?",
    "What is the standard SLA for billing complaints?",
    "Draft a brief 2-line apology for John.",
    "Should I escalate this to the senior billing team?",
    "What information should I collect from John before closing the ticket?"
]
conversation_log = []  

for question in email_questions:
    history_str= "\n".join([
        f"User: {q}\nAI: {a}" 
        for q, a in conversation_log
    ])
    if history_str:
        full_prompt = f"Previous conversation:\n{history_str}\n\nUser: {question}" 
    else:
        full_prompt = question

    response = llm.invoke(full_prompt).content
    conversation_log.append(question,response)
```
