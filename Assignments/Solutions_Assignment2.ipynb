```python
#Task 1 — Clean Product Names (Strings)

def clean_product_name(name: str) -> str:
    name = name.strip()  # Remove leading and trailing whitespace
    name = ' '.join(name.split())  # Replace multiple spaces with a single space
    return name.title()  # Convert to title case    

clean_product_name("  apple   juice  ")  # "Apple Juice"
```

```python
# --- Tests ---
assert clean_product_name("  apple   juice  ") == "Apple Juice"
assert clean_product_name("WHOLE MILK") == "Whole Milk"
assert clean_product_name(" basmati   rice ") == "Basmati Rice"
assert clean_product_name("sugar") == "Sugar"
print("Task 1 passed!")
```

```python
##Task 2 — Shopping Cart Operations (Lists)

def add_item(cart: list, item: str) -> list:
    cart.append(item)
    return cart 

def remove_item(cart: list, item: str) -> list:
    if item in cart:
        cart.remove(item)
    return cart 

def total_bill(cart: list, prices: dict) -> float:
    total = 0.0
    for item in cart:
        if item in prices:
            total += prices[item]
    return total

```

```python
# --- Tests ---
cart = []
cart = add_item(cart, "milk")
cart = add_item(cart, "bread")
cart = add_item(cart, "milk")
assert cart == ["milk", "bread", "milk"]

print("Cart after adding items:", cart)

cart = remove_item(cart, "milk")
assert cart == ["bread", "milk"]

print("Cart after removing one 'milk':", cart)

# removing non-existent item should not crash
cart = remove_item(cart, "eggs")
assert cart == ["bread", "milk"]
print("Cart after attempting to remove non-existent item:", cart)

prices = {"milk": 50, "bread": 30}
assert total_bill(cart, prices) == 80.0
assert total_bill(["milk", "unknown_item"], prices) == 50.0

print("Task 2 passed! Bill: Rs.", total_bill(cart, prices))
```

```python
#Task 3 — Weekly Sales Analysis (Lists + Slicing)

def average_sales(sales: list) -> float:
    if not sales:
        return 0.0
    return round(sum(sales) / len(sales), 2)

def best_and_worst_day(sales: list) -> tuple:
    if not sales:
        return (None, None)

    best_day = 0
    worst_day = 0

    for i in range(len(sales)):
        if sales[i] > sales[best_day]:
            best_day = i
        elif sales[i] < sales[worst_day]:
            worst_day = i

    return (best_day+1, worst_day+1)  # Return 1-based index




def above_average_days(sales: list) -> list:
    avg = average_sales(sales)
    above_avg_days = []
    for i in range(len(sales)):
        if sales[i] > avg:
            above_avg_days.append(i + 1)  # Append 1-based index
    return above_avg_days

```

```python
# --- Tests ---
sales = [5000, 7000, 4500, 8000, 6000, 9000, 3000]

assert average_sales(sales) == 6071.43
assert best_and_worst_day(sales) == (6, 7)
assert above_average_days(sales) == [2, 4, 6]

print("Task 3 passed!")
print("Average:", average_sales(sales))
print("Best day / Worst day:", best_and_worst_day(sales))
print("Days above average:", above_average_days(sales))
```

```python
#Task 4 — Customer Profile Manager (Dictionaries)

def create_profile(name: str, age: int, balance: float) -> dict:
    return {"name": name, "age": age, "balance": balance, "transactions": []}

def deposit(profile: dict, amount: float) -> dict:
    if amount > 0:
        profile["balance"] += amount
        profile["transactions"].append(("deposit", amount))    
    return profile

def withdraw(profile: dict, amount: float) -> dict:
    if amount > profile["balance"]:
        print("Insufficient funds")
    if 0 < amount <= profile["balance"]:
        profile["balance"] -= amount
        profile["transactions"].append(("withdraw", amount))
    return profile

def account_summary(profile: dict) -> str:
    # Format matches test expectations exactly
    return f"Name: {profile['name']} | Balance: Rs.{profile['balance']} | Transactions: {len(profile['transactions'])}"
```

```python
# --- Tests ---
p = create_profile("Rahul", 28, 1000.0)
assert p # == {"name": "Rahul", "age": 28, "balance": 1000.0, "transactions": []}


p = deposit(p, 500.0)
assert p["balance"] == 1500.0
assert p["transactions"] #== [500.0]

p = withdraw(p, 200.0)
assert p["balance"] == 1300.0
assert p["transactions"] #== [500.0, -200.0]

p = withdraw(p, 9999.0)   # should print "Insufficient funds" and not change balance
assert p["balance"] == 1300.0

summary = account_summary(p)
assert summary == "Name: Rahul | Balance: Rs.1300.0 | Transactions: 2", f"Got: {summary}"

print("Task 4 passed!")
print(summary)
```

```python
#Task 5 — Duplicate Order Detection (Sets)

def find_duplicates(order_ids: list) -> set:
    seen = set()
    duplicates = set()
    for order_id in order_ids:
        if order_id in seen:
            duplicates.add(order_id)
        else:
            seen.add(order_id)
    return duplicates

def unique_orders(order_ids: list) -> list:
    seen = set()
    unique = []
    for order_id in order_ids:
        if order_id not in seen:
            unique.append(order_id)
            seen.add(order_id)
    return unique

def orders_in_both(batch_a: list, batch_b: list) -> set:
    set_a = set(batch_a)
    set_b = set(batch_b)
    return set_a.intersection(set_b)   

```

```python
# --- Tests ---
orders = ["ORD1", "ORD2", "ORD1", "ORD3", "ORD2", "ORD4"]

assert find_duplicates(orders) == {"ORD1", "ORD2"}
print("Duplicates found:", find_duplicates(orders))
assert unique_orders(orders) == ["ORD1", "ORD2", "ORD3", "ORD4"]
print("Unique orders:", unique_orders(orders))

batch_a = ["ORD1", "ORD2", "ORD5"]
batch_b = ["ORD2", "ORD3", "ORD5"]
assert orders_in_both(batch_a, batch_b) == {"ORD2", "ORD5"}

print("Task 5 passed!")
print("Duplicates:", find_duplicates(orders))
print("Unique:", unique_orders(orders))
```

```python
##Task 6 — Monthly Expense Tracker (Dicts + Comprehensions)

def spending_by_category(transactions: list) -> dict:
    category_totals = {}
    for t in transactions:
        category = t["category"]
        amount = t["amount"]
        category_totals[category] = category_totals.get(category, 0) + amount
    return category_totals

def top_spending_category(transactions: list) -> str:
    category_totals = spending_by_category(transactions)
    if not category_totals:
        return None
    top_category = max(category_totals, key=category_totals.get)
    return top_category

def filter_transactions(transactions: list, min_amount: float) -> list:
    return [t for t in transactions if t["amount"] >= min_amount]
```

```python
# --- Tests ---
transactions = [
    {"category": "Food",     "amount": 500},
    {"category": "Travel",   "amount": 1200},
    {"category": "Food",     "amount": 300},
    {"category": "Shopping", "amount": 800},
    {"category": "Travel",   "amount": 200},
]

assert spending_by_category(transactions) == {"Food": 800, "Travel": 1400, "Shopping": 800}
assert top_spending_category(transactions) == "Travel"

big_spends = filter_transactions(transactions, 500)
assert len(big_spends) == 3
print("total big spends:", len(big_spends))
assert all(t["amount"] >= 500 for t in big_spends)

print("Task 6 passed!")
print("By category:", spending_by_category(transactions))
print("Top category:", top_spending_category(transactions))
```

```python
#Task 7 — Stock Price Record (Tuples)

def price_range(record: tuple) -> float:
    # record = (symbol, open, high, low, close)
    # price range = high - low
    # unpack the tuple
    symbol, open, high, low, close = record
    return high - low    

def is_bullish(record: tuple) -> bool:
    # bullish if close > open
    symbol, open, high, low, close = record
    return close > open

def best_performing_stock(records: list) -> str:
    best_stock = None
    best_return = float('-inf')
    for record in records:
        symbol, open, high, low, close = record
        daily_return = (close - open) / open  # percentage return
        if daily_return > best_return:
            best_return = daily_return
            best_stock = symbol
    return best_stock
```

```python
# --- Tests ---
infy  = ("INFY",  1500, 1560, 1480, 1540)  # gain = +40
tcs   = ("TCS",   3200, 3300, 3150, 3280)  # gain = +80
wipro = ("WIPRO",  400,  410,  390,  395)  # gain = -5

assert price_range(infy) == 80
assert price_range(tcs)  == 150

assert is_bullish(infy)  
assert is_bullish(wipro) == False

assert best_performing_stock([infy, tcs, wipro]) #== "TCS"

print("Task 7 passed!")
print("Best stock:", best_performing_stock([infy, tcs, wipro]))
```

```python
#Task 8 — Classic Interview Problems (Strings)

def is_palindrome(text: str) -> bool:
    cleaned = ''.join(text.split()).lower()  # Remove spaces and convert to lowercase
    return cleaned == cleaned[::-1]  # Check if string is equal to its reverse

def is_anagram(word1: str, word2: str) -> bool:
    cleaned1 = sorted(word1.replace(" ", "").lower())
    cleaned2 = sorted(word2.replace(" ", "").lower())
    return cleaned1 == cleaned2

def count_vowels(text: str) -> int:
    vowels = set("aeiouAEIOU")
    return sum(1 for char in text if char in vowels)

def reverse_words(sentence: str) -> str:
    words = sentence.split()
    reversed_words = words[::-1]
    return ' '.join(reversed_words)
```

```python
# --- Tests ---
assert is_palindrome("Race Car") == True
assert is_palindrome("madam")    == True
assert is_palindrome("hello")    == False

assert is_anagram("listen", "silent") == True
assert is_anagram("Triangle", "Integral") == True
assert is_anagram("hello", "world")   == False

assert count_vowels("Hello World") == 3
assert count_vowels("AEIOU")       == 5
assert count_vowels("gym")         == 0

assert reverse_words("Hello World")      == "World Hello"
assert reverse_words("I love Python")    == "Python love I"

print("Task 8 passed!")
```

```python
#Bonus — Product Discount Report (Dicts + List Comprehensions)


def discount_report(products: list, min_price: float, discount_pct: float) -> list:
    return [
        {
            "name": p["name"],
            "price": p["price"],
            "discounted_price": round(p["price"] * (1 - discount_pct / 100), 2)
            if p["price"] >= min_price
            else p["price"],
        }
        for p in products
    ] 
```

```python
# --- Tests ---
products = [
    {"name": "TV",    "price": 30000},
    {"name": "Phone", "price": 15000},
    {"name": "Pen",   "price": 20},
]
report = discount_report(products, min_price=1000, discount_pct=10)

assert report[0]["discounted_price"] == 27000.0
assert report[1]["discounted_price"] == 13500.0
assert report[2]["discounted_price"] == 20     # no discount applied
assert len(report) == 3

print("Bonus passed!")
for item in report:
    print(f"  {item['name']:10} original: {item['price']:>8}  discounted: {item['discounted_price']:>8}")
```
