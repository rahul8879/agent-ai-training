##TASK1: TYPE CASTING AND OPERATIONS

```python
def add_numeric_string(num_str: str, value: int) -> float:
    if type(num_str) != str:
        raise TypeError("num_str must be a string")
    else:
        return float(num_str) + value
```

```python
#Return the sum of the numeric string and Value as float.
Sum=add_numeric_string("10.5", 5)
print(Sum)

#Return the sum of the numeric string and Value as int
Sum=add_numeric_string("10", 5)
print(Sum)


#Return the sum of the numeric string and Value as String which would fail as expected. This need to be addressed using Try block.
Sum=add_numeric_string("abc", 5)
print(Sum)
```

Task 3: Keywords and Identifiers   

```python
def is_valid_identifier(name):
    # Check if the name is not empty
    if not name:
        return False
    # Check if the first character is a letter or underscore
    if not (name[0].isalpha() or name[0] == '_'):
        return False
    # Check if all characters are alphanumeric or underscores
    for char in name:
        if not (char.isalnum() or char == '_'):
            return False
    return True
```

```python
Result=is_valid_identifier("age1")  # True
print(Result)
Result=is_valid_identifier("1age")  # False
print(Result)
Result=is_valid_identifier("_age")  # True
print(Result)
```

Task4 : Types and Identity

```python
def compare_value_and_id(a, b) -> tuple:
    if a == b:
        same_value = True
    else:
        same_value = False
    if id(a) == id(b):
        same_id = True
    else:        
        same_id = False
    return (same_value, same_id)    
```

```python
result= compare_value_and_id("5", "5")  # (True, True)
print(result)

result= compare_value_and_id([1, 2], [1, 2])  # (True, False)
print(result)
```

```python
#Return Middle Element of the list.
def middle_element(values: list):
    if len(values) % 2 != 0:
        middle_index = len(values) // 2
        return values[middle_index]
    else:
        print ("The list has an even number of elements, no middle element exists.")
```

```python
Mid=middle_element([1, 2, 3, 4, 5])  # 3
print(Mid)

Mid=middle_element([1, 2, 3, 4])  # The list has an even number of elements, no middle element exists.
print(Mid)
```
